## Introduction

The dataset used in this project describes Euro daily exchange rates between 1999 and 2022. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

Daria Chemkaeva put together the data set and made it available on [Kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020?resource=download) — the data source is the European Central Bank. Note that the dataset gets regular updates.

```python
import pandas as pd
import matplotlib.pyplot as plt

exchange_rates = pd.read_csv("euro-daily-hist_1999_2022.csv")

display(exchange_rates.head(5))
display(exchange_rates.tail(5))
display(exchange_rates.info())
```

## Data Cleaning

The project will focus on the exchange rate between the Euro and the American dollar. Thus, the following steps are taken to clean the data:

- Rename the [US dollar ] and Period\Unit: columns to something easier to type:  `US_dollar` and `Time`.
- Change the `Time` column to a `datetime` data type.
- Sort the values by `Time` in ascending order.
- Reset the index (and drop the initial index).

```python
exchange_rates.rename(
    columns={"[US dollar ]": "US_dollar", "Period\\Unit:": "Time"}, inplace=True
)

exchange_rates["Time"] = pd.to_datetime(exchange_rates["Time"])
exchange_rates.sort_values("Time", inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)

euro_to_dollar = exchange_rates[
    ["Time", "US_dollar"]
]  # isolate the Time and US_dollar columns
euro_to_dollar["US_dollar"].value_counts()  # '-' character appears 62 times
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar["US_dollar"] != "-"]
euro_to_dollar["US_dollar"] = euro_to_dollar["US_dollar"].astype(float)
euro_to_dollar.info()
```

## Rolling Mean

```python
plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"])
plt.show()
```

The graph above appears jagged due to the number of data points it contains. A rolling mean can be used to show trends over time.

The rolling window can take various values — some common values include seven (one week), 30 (one month), 50 or 100 days, or 365 days (one year).

Below, we see how our line plot changes visually as we increase the rolling window. The window gets larger, and the line becomes smoother. This can be useful if we want the focus only on long-term trends.


```python
plt.figure(figsize=(9, 6))

plt.subplot(3, 2, 1)
plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"])
plt.title("Original", weight="bold")

for i, rolling_mean in zip([2, 3, 4, 5, 6], [7, 30, 50, 100, 365]):

    plt.subplot(3, 2, i)
    plt.plot(
        euro_to_dollar["Time"], euro_to_dollar["US_dollar"].rolling(rolling_mean).mean()
    )
    plt.title("Rolling Window:" + str(rolling_mean), weight="bold")

plt.tight_layout()
plt.show()
```

Calculate the rolling means for the US_dollar column using a moving window of 30 days. Add the rolling means to a new column named rolling_mean.

```python
euro_to_dollar["rolling_mean"] = euro_to_dollar["US_dollar"].rolling(30).mean()

euro_to_dollar
```

### Financial Crisis Visualization

```python
time_06_09 = euro_to_dollar.copy()[
    (euro_to_dollar["Time"].dt.year >= 2006) & (euro_to_dollar["Time"].dt.year <= 2009)
]
financial_crisis = euro_to_dollar.copy()[
    (euro_to_dollar["Time"].dt.year >= 2007) & (euro_to_dollar["Time"].dt.year <= 2008)
]
```

```python
import matplotlib.style as style

style.use("fivethirtyeight")

fig, ax = plt.subplots(figsize=(8, 3))

# Adding the plot using a rolling mean of 1 month (30 days)
ax.plot(time_06_09["Time"], time_06_09["rolling_mean"], linewidth=1)

# Highlighting the financial crisis (2007-2008)
ax.plot(
    financial_crisis["Time"], financial_crisis["rolling_mean"], linewidth=3, color="red"
)

# Highlight the crisis peak
ax.axvspan(xmin=13960, xmax=14125, ymin=0.09, alpha=0.3, color="grey")

# Define Tick Labels
ax.set_xticklabels([])
x = 13110
for year in ["2006", "2007", "2008", "2009", "2010"]:
    ax.text(x, 1.15, year, alpha=0.5, fontsize=11)
    x += 365

ax.set_yticklabels([])
y = 1.193
for rate in ["1.2", "1.3", "1.4", "1.5"]:
    ax.text(13000, y, rate, alpha=0.5, fontsize=11)
    y += 0.1

# Add title and subtitle
ax.text(
    13000,
    1.65,
    "Euro-USD rate peaked at 1.59 during 2007-2008 financial crisis",
    size=12,
    weight="bold",
)
ax.text(13000, 1.61, "Euro-USD exchange rate between 2006 and 2010", size=10)

ax.grid(alpha=0.3)
plt.show()
```
